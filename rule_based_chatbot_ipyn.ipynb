```python
def simple_chatbot():
  print("Chatbot:Hey!How may i help you?")

  while True:
    user_input = input("You:").lower()

    if user_input in["bye","exit","quit"]:
      print("Chatbot:Have a great day")
      break
    elif any(greeting in user_input for greeting in ["hi","hello","hey"]):
      print("Chatbot: Hello!I am here to help with your queries")
    elif "your name" in user_input:
      print("Chatbot:My name is Chatbot")
    elif 'how are you' in user_input:
      print("Chatbot:I'm just computer program thanks anyhow for asking")
    else:
      print("Chatbot:I'm sorry,I don't understand that")

if __name__ == "__main__":
    simple_chatbot()
```

```python
def physics_bot():
  physics_rule ={
      "physics": "The study of matter and energy",
      "force": "A push or pull that can change an object's motion",
      "gravity":"The attraction between two masses",
      "velocity": "The speed of an object in a specified direction",
      "energy": "The capacity is to do work"
  }
  print("PhysicsBot: Ask me a physics question (eg:what is force) or quit")
  while True:
    user_query= input("You:").lower()
    if user_query in ["quit","bye","exit"]:
      print("PhysicsBot:Goodbye")
      break
    found= False
    for key in physics_rule:
      if key in user_query:
        print(f"PhysicsBOT:{physics_rule[key]}")
        found = True
        break
    if not found:
      print("PhysicsBot:I don't know that one yet")
if __name__=="__main__":
  physics_bot()
```
